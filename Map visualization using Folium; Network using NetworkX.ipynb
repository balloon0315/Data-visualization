1.Load the housing_price.csv file and create the following map visualization with Folium.\\
a.The housing_price.csv spreadsheet include four columns: latitude, longitude, streetname, streetno, and price2014.
b.Place markers on a base map (OpenStreetMap). Each marker represents one house based on its latitude and longitude. 
  Each marker should be a circle filled with red color and with a black line. 
c.When the mouse cursor hovers over a marker, the streetno and streetname, and price2014 should be displayed in a tooltip.  

```python
import numpy as np
import pandas as pd
p= pd.read_csv("/Users/balloon_n/Documents/F/study/Data_Vis/proj4/Project4/Housing_price.csv")
hp=p[["latitude","longitude","streetname","streetno","price2014"]]
```

```python
hpt =hp.copy()
for col in hpt[["streetname","streetno","price2014"]]:
    hpt[col] = hpt[col].astype(str)
hpt["text"]= hpt["streetno"] + " "+ hpt["streetname"] + ", The Price is $" + hpt["price2014"]+ ". " 
```

```python
import folium
map1 = folium.Map(location=[42.3251,-72.6412],zoom_start=13,width=700,height=500)

locations = hpt[['latitude', 'longitude']]
locationlist = locations.values.tolist()
for i in range(0, len(locationlist)):
    folium.CircleMarker(locationlist[i],radius=5,tooltip = hpt["text"][i],
        fill=True,fill_color='red',color= 'black').add_to(map1) 
map1
```

2.Create a music collaboration network visualization using Plotly.  
Select 10 singers/musicians. For each musician, identify at least two collaborators and the songs they collaborated on. The more artists you include in your visualization the better. 

```python
import platform
import networkx as nx
from plotly.offline import download_plotlyjs, init_notebook_mode,iplot, plot
init_notebook_mode(connected=True)
import networkx as nx
```

```python
m= pd.read_csv("/Users/balloon_n/Documents/F/study/Data_Vis/proj4/Project4/musician.csv")
V=list(set().union(m.Musician,m.Collaborator))
E = list(zip(m["Musician"].tolist(), m["Collaborator"].tolist()))


g=nx.MultiGraph()
g.add_nodes_from(V)
g.add_edges_from(E)

def layouts (pos,V):
    Xn=[pos[k][0] for k in pos ]
    Yn=[pos[k][1] for k in pos ]
    trace_nodes=dict(type='scatter',
                 x=Xn,y=Yn,
                 mode='markers + text',
                 marker=dict(size=20, color="skyblue"),
                 text=V,
                 textposition="middle center",
                 hoverinfo='text')

    Xe=[]
    Ye=[]
    for e in g.edges():
        Xe.extend([pos[e[0]][0], pos[e[1]][0], None])
        Ye.extend([pos[e[0]][1], pos[e[1]][1], None])
    trace_edges=dict(type='scatter',
                     x=Xe,y=Ye,
                     mode="lines",
                     line=dict(width=1, color='grey'),
                     hoverinfo="none")
    layout=dict(width=850,height=850,autosize=False,showlegend=False,
                xaxis=dict(visible=False),yaxis=dict(visible=False),
               hovermode='closest')
    fig = dict(data=[trace_edges, trace_nodes], layout=layout)
    return fig

pos= nx.spring_layout(g,k=0.15,iterations=20)
fig1=layouts(pos,V)
iplot(fig1)
```

```python
posa= nx.shell_layout(g,scale=200)
fig2=layouts(posa,V)
iplot(fig2)
```

```python
posb=nx.fruchterman_reingold_layout(g) 
fig3=layouts(posb,V)
iplot(fig3)
```
